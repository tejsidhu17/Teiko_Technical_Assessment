```python
import pandas as pd

# Load CSV
df = pd.read_csv("cell-count.csv")

# Subject Level Columns
subject_level_cols = [
    "condition",
    "age",
    "sex",
    "treatment",
    "response"
]

# Group by subject and count unique values per column
variation_check = (
    df.groupby("subject")[subject_level_cols]
      .nunique()
)

# Find violations (any column with >1 unique value per subject)
violations = variation_check[variation_check.gt(1).any(axis=1)]

if violations.empty:
    print("All subject-level fields are invariant across samples.")
else:
    print("Some subject-level fields vary across samples.")
    print("\nSubjects with violations:")
    print(violations)

    print("\nDetailed breakdown:")
    for col in subject_level_cols:
        bad_subjects = variation_check[variation_check[col] > 1].index.tolist()
        if bad_subjects:
            print(f"  - {col} varies for subjects: {bad_subjects}")
```
